```python
import yaml
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

data_config = {
    'path': '/content/dataset',
    'train': 'images/train',
    'val': 'images/val',
    'nc': 1,
    'names': {0: 'plate'}
}

with open('/content/dataset/data.yaml', 'w') as f:
    yaml.dump(data_config, f)

```

```python
model.train(
    data='/content/dataset/data.yaml',
    epochs=50,
    imgsz=640,
    device='cuda'
)
```

```python
model.export(format='onnx')  # ou garder .pt
```
